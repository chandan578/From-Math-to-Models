## Diabates prediction using Decision Tree Regressor

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes
```

```python
diabetes['target']
```

```python
print(diabetes['DESCR'])
```

```python
## dependent feature
import pandas as pd
df_diabates = pd.DataFrame(diabetes['data'], columns=['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6'])
```

```python
x = df_diabates
x.head()
```

```python
## independent feature
y = diabetes['target']
y
```

```python
## train test split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)
```

```python
x_train.head()
```

```python
x_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
sns.heatmap(x_train.corr(), annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
decisionTree = DecisionTreeRegressor()
decisionTree.fit(x_train, y_train)
```

#### Hyperparameters tuning and model evaluation

```python
params = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error'],
    'splitter' : ['best', 'random'],
    'max_depth': [1,2,3,4,5,10,15, 20],
    'max_features': ['auto', 'sqrt', 'log2'] 
}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressor = DecisionTreeRegressor()
```

```python
grid = GridSearchCV(regressor, param_grid=params, cv=5, scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print(r2_score(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```

```python
selectmodel = DecisionTreeRegressor(criterion='squared_error', max_depth=3, max_features='log2', splitter='best')
```

```python
selectmodel.fit(x_train, y_train)
```

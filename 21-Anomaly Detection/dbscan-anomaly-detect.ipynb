## DBSCAN Anomaly Detection Implementation

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
```

```python
x, y = make_circles(n_samples=750, factor=0.3, noise=0.1)
```

```python
x
```

```python
plt.scatter(x[:,0], x[:, 1])
```

```python
plt.scatter(x[:,0], x[:, 1], c=y)
```

### DBSCAN

```python
dbscan = DBSCAN(eps=0.1)
```

```python
prediction = dbscan.fit_predict(x) ## which value is -1 then this is outlier
```

```python
prediction
```

```python
dbscan.labels_
```

```python
index = np.where(prediction < 0)
index
```

```python
plt.scatter(x[:,0], x[:, 1], c=dbscan.labels_)
plt.scatter(x[index,0], x[index, 1], c='r')  ## which dot is red that is our outlier value
```

## Local Outlier Factor

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
y_pred = lof.fit_predict(x)
```

```python
y_pred 
## 1 means inlier
## -1 means outlier
```

```python
scores = lof.negative_outlier_factor_
```

```python
scores
```

```python
index = np.where(y_pred < 0)
index
```

```python
plt.scatter(x[:,0], x[:, 1], c=y_pred)
plt.scatter(x[index,0], x[index,1], c='r')
```

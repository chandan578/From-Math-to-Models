## Isolation Forest Anomaly Detection

```python
import pandas as pd
```

```python
data = pd.read_csv('healthcare.csv')
```

```python
data
```

```python
import matplotlib.pyplot as plt
plt.scatter(data.iloc[:, 0], data.iloc[:, 1])
```

```python
from sklearn.ensemble import IsolationForest
```

```python
clf = IsolationForest(contamination=0.2)
clf.fit(data)
prediction = clf.predict(data)
```

```python
prediction
```

```python
import numpy as np
index = np.where(prediction < 0)
```

```python
index
```

```python
x = data.values
x
```

```python
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], edgecolor='r')
```

```python
plt.scatter(x[index, 0], x[index, 1], edgecolor='r')
```

```python
## now final result
plt.scatter(data.iloc[:, 0], data.iloc[:, 1])
plt.scatter(x[index,0], x[index,1], edgecolor='r')
```

## Logistic Regression Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
## create classification dataset
x, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)
```

```python
pd.DataFrame(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
x_train
```

### Model Training


```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
```

```python
logistic.fit(x_train, y_train)
```

```python
y_pred = logistic.predict(x_test)
print(y_pred)
```

```python
## performance metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_test, y_pred)
print(score)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

### Hyperparameter Tuning and Cross-Validation

#### Hyperparameter Tuning: 
 what is hyperparameter tuning?
- Hyperparameter tuning is the process of optimizing the hyperparameters of a machine learning model to improve its performance. Hyperparameters are settings that are not learned from the data but are set prior to training the model. Examples include learning rate, number of trees in a random forest, and regularization parameters.
- Techniques for hyperparameter tuning include grid search, random search, and Bayesian optimization.
#### Cross-Validation:
 what is cross-validation?
- Cross-validation is a technique used to assess how the results of a statistical analysis will generalize to an independent dataset. It is primarily used in settings where the goal is prediction and one wants to estimate how accurately a predictive model will perform in practice.
- The most common form of cross-validation is k-fold cross-validation, where the dataset is divided into k subsets, and the model is trained and validated k times, each time using a different subset as the validation set and the remaining subsets as the training set.


```python
model = LogisticRegression()
penalty = ['l1', 'l2', 'elasticnet']
c_value = [100, 10, 1.0, 0.1, 0.01]
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params = dict(penalty=penalty, C= c_value, solver = solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

### GridSerchCV

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv=cv, n_jobs=-1)
```

```python
grid
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(x_test)
```

```python
score = accuracy_score(y_test, y_pred)
print(score)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

### RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model = LogisticRegression()
randomCV = RandomizedSearchCV(estimator=model, param_distributions=params, cv=5, scoring='accuracy')
```

```python
randomCV.fit(x_train, y_train)
```

```python
randomCV.best_params_
```

```python
randomCV.best_score_
```

```python
y_pred = randomCV.predict(x_test)
```

```python
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### Logistic Regression with ROC curve and ROC AUC Score

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.linear_model import LogisticRegression
from matplotlib import pyplot
```

```python
## generate dataset
x, y = make_classification(n_samples=1000, n_classes=2, random_state=1)
```

```python
## split dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
## create a dummy model with only 0 output...
dummy_model_prob = [0 for _ in range(len(x_test))]
```

```python
dummy_model_prob
```

```python
## create a logistic model
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
## prediction based on probability
model_prob = model.predict_proba(x_test)
```

```python
model_prob
```

```python
## Lets focus on the positive outcome
model_prob = model_prob[:, 1]
model_prob
```

```python
## lets calculate the score
dummy_model_auc = roc_auc_score(y_test, dummy_model_prob)
model_auc = roc_auc_score(y_test, model_prob)
print(dummy_model_auc)
print(model_auc)
```

##### FPR is belogs to x axis and TPR belongs to y axis. 
--- If we plot then FPR is on x axis and TPR is on y axis.

```python
## Calculate ROC curve
dummy_fpr, dummy_tpr, dummy_thresh = roc_curve(y_test, dummy_model_prob)
model_fpr, model_tpr, model_thresh = roc_curve(y_test, model_prob)
```

```python
dummy_tpr
```

```python
## plot the roc curve for the model
pyplot.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy Model')
pyplot.plot(model_fpr, model_tpr, marker='*', label='Model')

pyplot.xlabel("False Positive Rate")
pyplot.ylabel("True Negative Rate")
pyplot.title('ROC curve for the Dummy vs Model')

pyplot.legend()
pyplot.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20, 50))   # use subplots instead of figure+add_subplot

# plot ROC curves on ax
ax.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy Model')
ax.plot(model_fpr, model_tpr, marker='.', label='Logistic')

# add threshold annotations
for fpr, tpr, thr in zip(model_fpr, model_tpr, model_thresh):
    ax.annotate(f'{thr:.2f}', xy=(fpr, tpr))

# axis labels
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')

# legend
ax.legend()

# show plot
plt.show()
```

## Tokenization Examples

Corpus : Corpus is a collection of written texts, often used for linguistic analysis or natural language processing tasks.
It can consist of books, articles, or any other form of written content.

Tokenization : Tokenization is the process of breaking down text into smaller units called tokens.
These tokens can be words, phrases, symbols, or other meaningful elements.
Types of tokenization include word tokenization, sentence tokenization, and subword tokenization.

```python
corpus = """Hello welcome, to my github repo.
Please visit all my repo and give star that ! and make fork also.
"""
```

```python
print(corpus)
```

```python
## Tokenization
## Paragraph --> Sentence
import nltk
from nltk.tokenize import sent_tokenize
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
document = sent_tokenize(corpus)
```

```python
document
```

```python
type(document)
```

```python
for sentence in document:
    print(sentence)
```

```python
## paragraph --> word
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
## sentence --> word
for sentence in document:
    print(word_tokenize(sentence))
```

```python
## word punctuation 
from nltk.tokenize import wordpunct_tokenize
```

```python
## paragraph --> word
## seprate word also punctuation
wordpunct_tokenize(corpus)
```

```python
## sentenct --> word
for sentence in document:
    print(wordpunct_tokenize(sentence))
```

```python
## small difference between those all
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
## in that it do not make seprate word for dot(.) , except for last
tokenizer.tokenize(corpus)
```

```python
for sentence in document:
    print(tokenizer.tokenize(sentence))
```

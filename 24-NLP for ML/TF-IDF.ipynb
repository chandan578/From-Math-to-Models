# TF-IDF 

TF-IDF stands for Term Frequency-Inverse Document Frequency. It is a numerical statistic that is intended to reflect how important a word is to a document in a collection or corpus. TF-IDF is commonly used in information retrieval and text mining.
### Components of TF-IDF
1. Term Frequency (TF): This measures how frequently a term appears in a document. The simplest way to calculate TF is to divide the number of times a term appears in a document by the total number of terms in that document.
   
   \[
   TF(t) = \frac{\text{Number of times term } t \text{ appears in a document}}{\text{Total number of terms in the document}}
   \]
2. Inverse Document Frequency (IDF): This measures how important a term is across all documents in the corpus. The IDF of a term is calculated by taking the logarithm of the total number of documents divided by the number of documents containing the term.

      \[
      IDF(t) = \log\left(\frac{\text{Total number of documents}}{\text{Number of documents containing term } t}\right)
      \]

```python
import pandas as pd

messages = pd.read_csv('sms-spam/spam.csv')
```

```python
messages = messages.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'])
messages
```

```python
messages.columns = ['label', 'message']
messages
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.stem import WordNetLemmatizer
import re
from nltk.corpus import stopwords

lemitizer = WordNetLemmatizer()
```

```python
corps = []

for i in range(len(messages)):

    ## first remove special character words
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower() ## lower each sentence
    review = review.split() ## convert to word
    review = [lemitizer.lemmatize(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corps.append(review)
```

```python
corps
```

## TF-IDF Calculation

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python

tfidf = TfidfVectorizer(max_features=100)
x = tfidf.fit_transform(corps).toarray()
tfidf.vocabulary_
```

```python
## now using n-grams
tfidf = TfidfVectorizer(max_features=200, ngram_range=(1,2))
x = tfidf.fit_transform(corps).toarray()
tfidf.vocabulary_
```
